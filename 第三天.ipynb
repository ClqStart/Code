```python
df_data_left=data[["lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5"]]
df_data_right=data[["rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]]
df_data_left.columns=["pickhero1","pickhero2","pickhero3","pickhero4","pickhero5"]
df_data_right.columns=["pickhero1","pickhero2","pickhero3","pickhero4","pickhero5"]
df_data_last=pd.concat((df_data_left,df_data_right))
```

```python
df_data_last.head()
```

```python
df_data_last.columns=['top','junkle','mid','ad','support']
df_data_last.head()
```

```python
top =df_data_last.groupby('top').size().count()
junkle=df_data_last.groupby('junkle').size().count()
mid =df_data_last.groupby('mid').size().count()
ad =df_data_last.groupby('ad').size().count()
support =df_data_last.groupby('support').size().count()
display(top,junkle,mid,ad,support)
```

```python
pie_data = [top, junkle, mid, ad, support]
pie_lables = ["上单", "打野", "中单", "射手", "辅助"]
pie_colors = ["r", "g", "b", "y", "cyan"]
pie_explode = [0, 0, 0.1, 0, 0.2]

# 引入matplotlib的是， 给pyplot对象起的一个别名， 这个对象就是一个绘图对象
plt.pie(pie_data, labels=pie_lables, colors=pie_colors, autopct="%.2f%%",explode=pie_explode)
```
