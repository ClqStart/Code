# 分析需求2：胜场最多的10位英雄

这个需求有一个难点：就是每一条数据中，两支队伍进行一场比赛，那到底是谁赢了，是由数据中的 result 这个字段告诉我我们的。result字段的结果只有两种状态：r和l，表示right和left，也就是到底是那个队伍赢的。right team还是 left team

```python
df["result"].head()
```

那么为了方便统计，需要把数据的位置进行调换。

```python
# 胜负数据，10个英雄
columns = ["result", "lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5","rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]

# 获取胜利的英雄
data = df[columns]

# 进行筛选，把整个dataframe中的 result为 l， "lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5"
data_left = data[data["result"] == "l"][["lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5"]]
data_right = data[data["result"] == "r"][["rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]]
display(data_left, data_right)
```

```python
display(data_left.columns, data_right.columns)
```

```python
# 因为左边队伍和右边队伍的pick英雄字段名不一致，所以调整成为一致的
data_left.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_right.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]

# 这一句代码是进行两份数据的叠加整合
data_last = pd.concat((data_left, data_right))
display(data_last)
```

```python
# 统计每个英雄的胜场次数
victory_hero = data_last.values.flatten()
result_df = pd.DataFrame(victory_hero, columns=["name"])
victory_hero = result_df["name"].value_counts()

# 获取胜场最多的十个英雄
top10_victory_heros = victory_hero.head(10)
display(top10_victory_heros)

# 英雄的胜率： 159/283 =  54%
```

# 分析需求3：分析所有英雄的胜率，并取出前10进行图表展示

胜率最高 和 胜场最多 是两个不同的概念。    
胜率 = 胜场 / 总出场次数

这个需求有一个难点：就是每一条数据中，两支队伍进行一场比赛，那到底是谁赢了，是由数据中的 result 这个字段告诉我我们的。result字段的结果只有两种状态：r和l，表示right和left，也就是到底是那个队伍赢的。right team还是 left team

在上面的需求中，已经能把每个英雄的胜场求出来了， 那么同样的，也可以把每个英雄的败场次数求出来

```python
# 英雄列
columns = ["result", "lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5","rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]

# 获取比赛胜利的英雄
data = df[columns]
data_left_victory = data[data["result"] == "l"][["lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5"]]
data_right_victory = data[data["result"] == "r"][["rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]]

data_left_victory.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_right_victory.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_victory_last = pd.concat((data_left_victory, data_right_victory))

# 统计每个英雄的胜场次数
victory_heros = data_victory_last.values.flatten()
victory_hero_df = pd.DataFrame(victory_heros, columns=["victory"])
victory_heros = victory_hero_df["victory"].value_counts()

display(victory_heros)
```

```python
# 获取比赛失败的英雄
data_left_failure = data[data["result"] == "r"][["lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5"]]
data_right_failure = data[data["result"] == "l"][["rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]]

data_left_failure.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_right_failure.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_failure_last = pd.concat((data_left_failure, data_right_failure))

# 统计每个英雄的胜场次数
failure_heros = data_failure_last.values.flatten()
failure_hero_df = pd.DataFrame(failure_heros, columns=["failure"])
failure_heros = failure_hero_df["failure"].value_counts()

display(failure_heros)

# 总共：卡莎;283场    赢：159次，  输：124
# 胜率：  159/283
```

```python
last_result_data = failure_heros + victory_heros
display(last_result_data)
```

```python
# display(victory_heros.head(), failure_heros.head())
# last_result = victory_heros / (victory_heros + failure_heros)
# display(last_result)

# 合并英雄胜利和失败的数据
result_df = pd.concat((victory_heros, failure_heros), axis=1)

# 填充默认值
result_df.fillna(0, inplace=True)

# 计算胜率
result_df["vict_percent"] = result_df["victory"] / (result_df["victory"] + result_df["failure"])
sort_result = result_df.sort_values("vict_percent", ascending=False)

# 展示结果
display(sort_result.head(30))


# 图表展示
sort_result["vict_percent"].head(20).plot(kind="bar")
```

# 分析需求4：每个位置的出场英雄数饼图分析

简单来说，就是统计，每个位置上，出现的英雄的总数量，然后绘饼图来展示

```python
# 英雄列
columns = ["result", "lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5","rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]

# 获取比赛胜利的英雄
data = df[columns]
data_left = data[["lpickhero1","lpickhero2","lpickhero3","lpickhero4","lpickhero5"]]
data_right = data[["rpickhero1","rpickhero2","rpickhero3","rpickhero4","rpickhero5"]]

# 位置：上单，打野，中单，射手，辅助
data_left.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_right.columns = ["pickhero1", "pickhero2", "pickhero3", "pickhero4", "pickhero5"]
data_last = pd.concat((data_left, data_right), axis=0)


# 获取五个位置的总出场英雄数
top = data_last.groupby("pickhero1").size().count()    # 获取出现在上单位置上的总英雄个数
jungle = data_last.groupby("pickhero2").size().count()
mid = data_last.groupby("pickhero3").size().count()
ad = data_last.groupby("pickhero4").size().count()
support = data_last.groupby("pickhero5").size().count()

display(top, jungle, mid, ad, support)


# a,b,a
# b,c,d
# c,d,a
# a,b,e
# c,a,b

groupby()
([a,a], [b], [c,c])


size()
([a,2], [b,1], [c,2])

count()
3


# 其他的聚合操作：
# count  sum  max  min   mean ....


# 绘制饼图
```

## 饼图
plt.pie 饼图
* labels 每个部分显示的标签。
* explode 指定每个部分距离圆心的偏移量（单位为半径的长度）。
* colors 指定每个部分的颜色。
* autopct 设置每个部分显示的比例值（格式化）。
* couterclock 是否逆时针绘图。默认为True。
* startangle 初始绘图点位置（逆时针偏移x轴的角度）。默认为偏移0度（x轴）。
* shadow 是否含有阴影，默认为False。

```python
pie_data = [top, jungle, mid, ad, support]
pie_lables = ["上单", "打野", "中单", "射手", "辅助"]
pie_colors = ["r", "g", "b", "y", "cyan"]
pie_explode = [0, 0, 0.1, 0, 0.2]

# 引入matplotlib的是， 给pyplot对象起的一个别名， 这个对象就是一个绘图对象
plt.pie(pie_data, labels=pie_lables, colors=pie_colors, autopct="%.2f%%")
```
